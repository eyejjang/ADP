```python
# feature 삭제
del_feat = ['PassengerId', 'Cabin']
titanic = titanic.drop(del_feat, axis=1)

# 질적변수의 데이터 타입을 category로 변경
qual = ['Survived', 'Pclass', 'Name', 'Sex', 'Ticket', 'Embarked']
quan = ['Age', 'SibSp', 'Parch', 'Fare']
for col in qual:
    titanic[col] = titanic[col].astype('category')
print(titanic.info())
```

## 1-2. 질적변수
### - 도수분포표, 상대도수분포표

```python
# 도수분포표
from pandas import DataFrame
#freq = DataFrame(titanic['Survived'].value_counts())
#print(freq)
freq = DataFrame(titanic['Survived'].value_counts()).reset_index(drop=True)
freq = freq.rename(columns={'count':'Survived'})
print(freq)
```

```python
# 상대도수분포표 추가
freq['상대도수'] = titanic['Survived'].value_counts(normalize=True)
freq = freq.rename(columns={'Survived':'도수'}) #'Survived'를 '도수'로 컬럼명 변경
print(freq)
```

### - 상관계수: 스피어만 상관계수, 켄달의 타우

```python
from scipy.stats import spearmanr, kendalltau
corr, p = spearmanr(titanic['Survived'], titanic['Pclass'])
print("스피어만 상관계수는 {:.3f}, p-value는 {:.3f}".format(corr, p))

corr2, p2 = kendalltau(titanic['Survived'], titanic['Pclass'])
print("켄달타우 상관계수는 {:.3f}, p-value는 {:.3f}".format(corr2, p2))
```

## 1-3. 양적변수
### - 도수분포표, 상대도수분포표

```python
# 계급의 개수를 지정하여 도수분포표 생성
from pandas import cut
result1 = cut(titanic['Age'], bins=5)
freq1 = DataFrame(result1.value_counts()).reset_index()
# 컬럼명을 변경하고, 계급 오름차순 정렬
#freq1 = freq1.rename(columns={'index':'계급', 'Age':'도수'}).sort_values(by='계급')
freq1 = freq1.rename(columns={'Age':'계급', 'count':'도수'}).sort_values(by='계급')

# NaN 처리 필요
# result2 = DataFrame(result1)
# freq1['상대도수'] = result2['Age'].value_counts(normalize=True)
# print(freq1)

# freq1의 상대도수 추가 (freq1의 각 도수를 전체 도수의 합으로 나누어 구함)
freq1['상대도수'] = freq1['도수'].apply(lambda x: x / freq1['도수'].sum())
print(freq1)
```

### - 왜도, 첨도

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.skew.html

#### scipy.stats. 
## skew

**Notes**
***

The sample skewness is computed as the Fisher-Pearson coefficient of skewness,   
i.e.  
$$
g_1=\frac{m_3}{m_2^{3 / 2}}
$$

where  

$$
m_i=\frac{1}{N} \sum_{n=1}^N(x[n]-\bar{x})^i
$$

is the biased sample $i$th central moment, and $\bar{x}$ is the sample mean. If bias is False, the calculations are corrected for bias and the value computed is the adjusted Fisher-Pearson standardized moment coefficient,   
i.e.

$$
G_1=\frac{k_3}{k_2^{3 / 2}}=\frac{\sqrt{N(N-1)}}{N-2} \frac{m_3}{m_2^{3 / 2}} .
$$

* IMG -> LaTex 변환 : https://snip.mathpix.com/home

```python
# pandas
print("왜도: %.3f, 첨도: %.3f" %(titanic['Fare'].skew(), titanic['Fare'].kurt()))

# scipy
from scipy.stats import skew, kurtosis
ssF = skew(titanic['Fare'], bias=False)
skF = kurtosis(titanic['Fare'], bias=False)
ssT = skew(titanic['Fare'], bias=True)
skT = kurtosis(titanic['Fare'], bias=True)
print("왜도: %.3f(%.3f), 첨도: %.3f(%.3f)" %(ssF, ssT, skF, skT))
```
